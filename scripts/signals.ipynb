# Signals
1. Plotting of continuous-time signals.
    1. Compute the signal

        $$
        x_1(t)=5\sin{12t}
        $$

        at 500 points in the time interval $0\leq t\leq 5\ s$ and graph the result.

```python
t = np.linspace(0, 5, 500)

x1 = 5*np.sin(12*t)

plt.plot(t, x1)
plt.grid()
plt.show()
```
